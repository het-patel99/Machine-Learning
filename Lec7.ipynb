```python
import random
def load_dataset(filename, split, training_set=[], testing_set=[]):
    with open(filename,'r') as csvfile:
        lines = csv.reader(csvfile)
        dataset = list(lines)
        for x in range(len(dataset)-1):
            for y in range(9):
                dataset[x][y] = float(dataset[x][y])
            if random.random() < split:
                training_set.append(dataset[x])
            else:
                testing_set.append(dataset[x]) 
```

```python
# Calculating Similarity
import math
def euclidean_distance(ins1, ins2, length):
    distance = 0
    for x in range(length):
        distance += pow((ins1[x] - ins2[x]),2)
    return math.sqrt(distance)
```

```python
# getting nearest neighbors
import operator

def get_neighbors(training_set,test_instance,k):
    distances = []
    length = len(test_instance) - 1
    for x in range(len(training_set)):
        dist = euclidean_distance(test_instance, training_set[x], length)
        distances.append((training_set[x],dist))
    distances.sort(key=operator.itemgetter(1))
    neighbors = []
    for x in range(k):
        neighbors.append(distances[x][0])
    return neighbors
```

```python
import operator

def get_response(neighbors):
    classVotes = {}
    for x in range(len(neighbors)):
        response = neighbors[x][-1]
        if response in classVotes:
            classVotes[response] += 1
        else:
            classVotes[response] = 1
    #sortedVotes = classVotes.sort(key=operator.itemgetter(1), reverse = True)
    sortedVotes = sorted(classVotes.items(), key=operator.itemgetter(1), reverse = True)
    return sortedVotes[0][0]
```

```python
# Accuray 

def get_accuracy(testing_set, predictions):
    true = 0
    for x in range(len(testing_set)):
        if testing_set[x][-1] == predictions[x]:
            true+=1
    return (true/float(len(testing_set)))*100.0
```

```python
# Defining main function

def main():
    training_set = []
    testing_set = []
    split = 0.80
    load_dataset(r'Desktop//dataset//cmc.data',split,training_set,testing_set)
    print('No. of Training data : ' + repr(len(training_set)))
    print('No. of Testing data : ' + repr(len(testing_set)))
    predictions = []
    k = 8
    for x in range(len(testing_set)):
        neighbors = get_neighbors(training_set,testing_set[x],k)
        result = get_response(neighbors)
        predictions.append(result)
        print(' Predicted Value = '+ repr(result) + ', Actual value = '+ repr(testing_set[x][-1]))
    accuracy = get_accuracy(testing_set,predictions)
    print('Accuracy : ' + repr(accuracy) + '%')
main()
```
